## Problem Statement

In a bid to improve air quality, vehicle manufacturers and policy makers are working to reduce carbon emissions from road vehicles by increasing availability of electric vehicles (EVs) and EV charging stations. Policy makers are aiming to meet net zero emission targets, and have requested analysis into the progress and effectiveness of building out EV infrastructure and increasing EV registration on air pollution levels. As part of progressing towards net zero emissions, policy makers are seeking to determine how many EV registrations and charging stations will yield a 10% improvement to the air quality index. This project analyzes EV trends and data collected on air quality, and uses a time-series model to predict air quality based on EV registration and charging station quantities.

## Background

Air pollution has become a major environmental issue worldwide, contributing to several respiratory and cardiovascular diseases [[source]](https://www.transportation.gov/mission/health/cleaner-air). Motor vehicles have been identified as one of the [leading sources of air pollution](https://www.who.int/health-topics/air-pollution#tab=tab_1), which has led to efforts aimed at reducing their carbon emissions. One approach to achieving this goal is through the promotion of electric vehicles (EVs) and the expansion of EV charging stations. In 2021, President Biden even signed an executive order that set an ambitious target of having EVs account for 50% of all new car sales by the year 2030 [[source]](https://www.whitehouse.gov/briefing-room/statements-releases/2021/08/05/fact-sheet-president-biden-announces-steps-to-drive-american-leadership-forward-on-clean-cars-and-trucks/). This strategy has continued to gain traction in recent years, with manufacturers themselves making [pledges](https://www.caranddriver.com/news/g35562831/ev-plans-automakers-timeline/) to build out their EV lineups, providing better, more affordable, and simply *more* EV options to consumers.

Despite these efforts, which include the promise of up to [$7,500 in tax credits](https://www.irs.gov/credits-deductions/credits-for-new-clean-vehicles-purchased-in-2023-or-after) for EVs purchased in 2023 or later, [a poll of Americans](https://www.pbs.org/newshour/nation/many-americans-arent-yet-sold-on-going-electric-for-their-next-car-poll-shows) revealed that many consumers are still reluctant to make the switch to an EV. There are a variety of reasons, but of those polled who said they are not likely to purchase an EV, 3 out of 4 cited an insuffienct number of charging stations as a reason why. With the number of EV charging stations serving as a limiting factor for EV sales, the Biden administration has also set its sights on building out an expansive network of EV chargers across America [[source]](https://www.whitehouse.gov/briefing-room/statements-releases/2023/02/15/fact-sheet-biden-harris-administration-announces-new-standards-and-major-progress-for-a-made-in-america-national-network-of-electric-vehicle-chargers/). To quantify the impact of increasing EV sales and the expansion of the nation's EV charging network, this analysis will consider air quality index data ranging from the year 2000-2022, charging station data from 2007-2022, and electric vehicle data from 2016-2021.

## Data Import & Cleaning

### Dataset 1: Annual Air Quality Index by County, 2000 - 2022
Data source: United States Environmental Protection Agency, https://aqs.epa.gov/aqsweb/airdata/download_files.html

```python
# Create a files variable that contains all of our data files.
files = os.listdir("../data")
```

```python
files[:]
```

```python
# Read in data from 2000
year_2000 = pd.read_csv("../data/annual_aqi_by_county_2000.csv")
year_2000.head()
```

```python
# function to read data
def process_data(file):
    df = pd.read_csv("../data/" + file)
    return df
```

```python
file_list = [file for file in files if file.startswith("annual_aqi_by_county_")]
file_list = [process_data(file) for file in file_list]
```

```python
air_quality = pd.concat(file_list, axis=0)
```

```python
air_quality.columns = air_quality.columns.str.lower().str.replace(' ', '_', regex=False)
```

```python
# info
air_quality.info()
```

```python
# check null value
air_quality.isna().sum()
```

```python
air_quality.describe()
```

```python
air_quality = air_quality[air_quality['max_aqi'] <= 500]
```

```python
air_quality['state'].unique()
```

```python
air_quality = air_quality[~air_quality['state'].isin(['Canada', 'Country Of Mexico', 'Puerto Rico', 'Virgin Islands'])]
```

```python
quality_cols = [
    'good_days', 
    'moderate_days',
    'unhealthy_for_sensitive_groups_days',
    'unhealthy_days',
    'very_unhealthy_days',
    'hazardous_days'
]

cols_to_sum = [col for col in quality_cols if col != 'good_days']

# Create the new column
air_quality['bad_days'] = air_quality[cols_to_sum].sum(axis=1)
```

```python
day_count_cols = [
    'good_days',
    'moderate_days',
    'unhealthy_for_sensitive_groups_days',
    'unhealthy_days', 'very_unhealthy_days',
    'hazardous_days',
    'days_co',
    'days_no2',
    'days_ozone',
    'days_pm2.5',
    'days_pm10',
    'bad_days'
]

# iterate through the columns and divide each by 'days_with_aqi'
for col in day_count_cols:
    air_quality[f'pct_{col}'] = air_quality[col] / air_quality['days_with_aqi']
```

```python
air_quality = air_quality.drop(columns=day_count_cols)
```

```python
air_quality.head()
```

```python
air_quality.describe()
```

```python
# save csv file
air_quality.to_csv('../output/air_quality_by_county.csv', index=False)
```

```python
air_quality_by_state = air_quality.groupby(['state', 'year']).mean(numeric_only=True).reset_index()
air_quality_by_state
```

```python
# save csv file
air_quality_by_state.to_csv('../output/air_quality_by_state.csv', index=False)
```

### Dataset 3: Alternative Fueling Station Counts by State, 2007 - 2022
Data source: U.S. Department of Energy, https://afdc.energy.gov/stations/states

```python
# read data from excel_sheet to check the datafram information
station_counts_2022 = pd.read_excel("../data/historical-station-counts.xlsx", sheet_name="2022")
station_counts_2022.head(20)
```

```python
# setup path
file = "../data/historical-station-counts.xlsx"
```

```python
# check the sheet_names
my_sheets = pd.ExcelFile(file)
my_sheets.sheet_names
```

```python
# read each sheet into a datafram and add all df into one big datafram
xls = pd.ExcelFile('../data/historical-station-counts.xlsx')

station_counts = pd.DataFrame()
for sheet in xls.sheet_names:
    df = pd.read_excel(xls, sheet)
    df['year'] = sheet
    station_counts = pd.concat([station_counts, df])
```

```python
station_counts.head()
```

```python
# Drop unnecessary columns
# station_counts.drop(["Biodiesel", "CNG", "E85", "Electric", "Hydrogen", "LNG", "Propane"], axis=1, inplace=True)
```

```python
# drop the row with nan due to original data formating
station_counts.dropna(subset=['State'], inplace=True)

# drop the row with "Total" for the column of State
station_counts.drop(station_counts[station_counts['State']== 'Total'].index  , inplace=True)
```

```python
# check the value in the column of State
station_counts['State'].value_counts()
```

```python
station_counts.isna().sum()
```

```python
station_counts.info()
```

```python
# save to csv file
station_counts.to_csv("../output/station_counts_by_state.csv", index=False)
```

```python
station_counts.head()
```

```python
# extrat the total charging station count from Electric column
station_counts['Electric'] = station_counts['Electric'].apply(lambda s: re.findall(r'\S+', str(s))[-1])

# extrat the total charging station count from Hydrogen column
station_counts['Hydrogen'] = station_counts['Hydrogen'].apply(lambda s: re.findall(r'\S+', str(s))[-1])

# extrat the total charging station count from Propane column
station_counts['Propane'] = station_counts['Propane'].apply(lambda s: re.findall(r'\S+', str(s))[-1])
```
